5 features into two classes neural network


```python
import torch
import torch.nn as nn
```

```python
 # inherting from the nn.Module class
class Model(nn.Module):
  def __init__(self,num_features):
    super().__init__()# invoking the constructor
    self.linear = nn.Linear(num_features,1) # one neuron
    self.sigmoid  = nn.Sigmoid() # sigmoid activation

  def forward(self,features):
    out = self.linear(features) # z = wx+b
    out = self.sigmoid(out)
    return out
```

```python
features = torch.rand(10,5)
```

```python
model = Model(features.shape[1])
model(features)
```

Better neural network

```python
# 5 features , 3 neurons(relu) , one output layer(sigmoid)

class Model(nn.Module):
  def __init__(self,num_features):
    super().__init__()
    self.linear1 = nn.Linear(num_features,3)
    self.relu = nn.ReLU()
    self.linear2 = nn.Linear(3,1)
    self.sigmoid  = nn.Sigmoid()

  def forward(self,features):
    out = self.linear1(features)
    out = self.relu(out)
    out = self.linear2(out)
    out = self.sigmoid(out)
    return out
```

```python
model = Model(features.shape[1])
model(features)
```

```python
class Model(nn.Module):
  def __init__(self,num_features):
    super().__init__()
    self.network = nn.Sequential(
          nn.Linear(num_features,3),
          nn.ReLU(),
          nn.Linear(3,1),
          nn.Sigmoid()
    )

  def forward(self,features):
    out = self.network(features)
    return out
```

```python
model = Model(features.shape[1])
model(features)
```

```python
summary(model,input_size=(10,5))
```
