```python
import tensorflow as tf
from tensorflow import keras
```

```python
from keras.datasets import imdb
from keras import Sequential
from keras.layers import Flatten,SimpleRNN,Dense,Embedding
```

```python
(X_train,y_train),(X_test,y_test) = imdb.load_data()
```

# Integer Encoding -
 where each word in the dataset will get a index integer value and so that every sentence is of equal length we add padding of 0

```python
print(X_train[0])
```

```python
from keras.utils import pad_sequences
# currently trimming the entire dataset to acquire only 50 words per document
# will try with full later
X_train = pad_sequences(X_train,padding ='post',maxlen=50)
X_test = pad_sequences(X_test,padding ='post',maxlen=50)
```

```python
X_train.shape
```

return sequences = false means the output of every node and the end of a t state will be in the memory itself and will not be returned. But this will be returned to us if stated True

```python
model = Sequential()
model.add(SimpleRNN(32,input_shape=(50,1),return_sequences=False))
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
history = model.fit(X_train,y_train,epochs=10,validation_data=(X_test,y_test))
```
