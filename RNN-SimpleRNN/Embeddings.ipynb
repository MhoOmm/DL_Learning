```python
import tensorflow
from tensorflow import keras
from keras.datasets import imdb
```

```python
(X_train,y_train),(X_test,y_test) = imdb.load_data()
```

Embedding Techniques

```python
from keras import Sequential
from keras.layers import Input,SimpleRNN,Dense,Flatten,Embedding
```

```python
X_train.shape
```

```python
from keras.utils import pad_sequences
# currently trimming the entire dataset to acquire only 50 words per document
# will try with full later
X_train = pad_sequences(X_train,padding ='post',maxlen=50)
X_test = pad_sequences(X_test,padding ='post',maxlen=50)
```

```python
X_train.shape
```

```python
vocab_size = int(X_train.max()) + 1
```

```python
model = Sequential()
# Embedding = (input= no of unique words in the vocab ,output= what is the size of the dense vector to be returned,input_length= no of words)
model.add(Input(shape=(50,)))
model.add(Embedding(input_dim=vocab_size, output_dim=2)) # this will create a vector of 2 for every word in the dataset
model.add(SimpleRNN(32,return_sequences=False))
model.add(Dense(1,activation='sigmoid'))

model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

model.summary()
```

```python
history = model.fit(X_train,y_train,epochs=10,validation_data=(X_test,y_test))
```
